```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import random
import copy
import matplotlib.pyplot as plt
```

```python
class Set_Attributes():
    def __init__(self,number_of_attributes):
        self.number_of_attributes = number_of_attributes
        attributes_names = []
        for num in range(number_of_attributes):
            attributes_names.append('attribute_' + str(num+1))
        self.attributes_names = attributes_names
        self.card_deck = self.generate_card_set()
        self.caps = [1,2,4,9,20,45,112]


    def generate_card_set(self):
        card_deck = pd.DataFrame(columns = self.attributes_names)
        attributes =[[]]*self.number_of_attributes
        options = [0,1,2]
        attributes[0] = [0,1,2]
        for position in range(1,len(attributes)):
            attributes[position] = []
            for option in options:
                for card in attributes[position-1]:
                    new_card = np.array(card)
                    new_card = np.append(new_card,option)
                    attributes[position].append(new_card)

        attributes  = np.array(attributes[-1])
        for att in range(self.number_of_attributes):
            card_deck[card_deck.columns[att]] = attributes[:,att]
            card_deck.iloc[:,att] = attributes[:,att]

        self.card_deck = card_deck
        return self.card_deck

    def generate_types_of_sets(self):
        set_type=pd.DataFrame(columns = self.attributes_names)
        set_type['number_of_sets'] = []
        types_of_sets = []

        def per_type (bs = []):
            if len(bs) == self.number_of_attributes:
                types_of_sets.append(bs)
            else:
                per_type(bs + ['same'] )
                per_type(bs + ['different'])

        per_type()
        types_of_sets = np.array(types_of_sets[1:])
        num = []
        for itr in range(len(types_of_sets)):
            count = np.sum(types_of_sets[itr] == 'different')
            possibilities = (3**self.number_of_attributes)/6
            num.append(possibilities*(2**count))

        for att in range(self.number_of_attributes):
            set_type[set_type.columns[att]] = types_of_sets[:,att]

        set_type['number_of_sets'] = num
        return set_type


    def check_set(self, card_1,card_2,card_3):
        cards = [card_1,card_2,card_3]
        cards = (sorted(cards))
        attributes =[]

        for att in range(self.number_of_attributes):
            per_att = []
            for card in cards:
                per_att.append(self.card_deck.iloc[card][self.attributes_names[att]])
            attributes.append(per_att)

        for single_att in attributes:
            if len(np.unique(single_att)) == 2:
                return False

        return True

    def generate_list(self,current_board):
        possible_sets = []
        for i in range(len(current_board)):
            for j in range(i+1,len(current_board)):
                for k in range(j+1,len(current_board)):
                    possible_sets.append([current_board[i],current_board[j],current_board[k]])
        self.possible_sets = possible_sets

  # def check_list(self):
  #   for i in self.possible_sets:
  #       value = self.check_set(i[0],i[1],i[2])
  #       if value == True:
  #         current_board.remove(i)
  #         current_board.remove(j)
  #         current_board.remove(k)
  #         break
  #   return current_board,value

    def generate_twos(self,current_board,k):
        possible_sets = []
        for i in range(len(current_board)):
            for j in range(i+1,len(current_board)):
                possible_sets.append([current_board[i],current_board[j],k])
        self.possible_sets = possible_sets

    def check_minimum_number(self,card_deck,deck_cards,current_board):
        new_addition = random.sample(deck_cards,1)
        current_board.append(new_addition[0])
        deck_cards.remove(new_addition[0])
        self.generate_twos(current_board,new_addition[0])
        count_sets = 0
        for i in self.possible_sets:
            value = self.check_set(i[0],i[1],i[2])
            if value == True:
                return (len(current_board))

        b = self.check_minimum_number(card_deck,deck_cards,current_board)
        return b
    
    def run_check(self):
        deck_cards = [*range(0,len(self.card_deck))]
        current_board = random.sample(deck_cards,2)
        deck_cards.remove(current_board[0])
        deck_cards.remove(current_board[1])
        while len(deck_cards) > 0:
            new_addition = random.choice(deck_cards)
            self.generate_twos(current_board,new_addition)
            check_value = False
            for i in self.possible_sets:
                value = self.check_set(i[0],i[1],i[2])
                check_value = check_value + value
                if value == True:
                    break
            if check_value == 0:
                current_board.append(new_addition)
            deck_cards.remove(new_addition)
            
        return current_board
```
