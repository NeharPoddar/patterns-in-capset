```python
def check_set(card_1,card_2,card_3):
    # print(card_deck.loc[[card_1,card_2,card_3],:])
    cards = [card_1,card_2,card_3]
    cards = (sorted(cards))
    shape = []
    color = []
    pattern = []
    number = []
    for card in cards:
        shape.append(card_deck['Shape'][card])
        color.append(card_deck['Color'][card])
        pattern.append(card_deck['Pattern'][card])
        number.append(card_deck['Number'][card])
    if len(np.unique(shape)) == 2 or len(np.unique(color)) == 2 or len(np.unique(pattern)) == 2 or len(np.unique(number)) == 2:
        # print('NOT A SET')
        return False
    else:
        # print('IT IS A SET')
        # print(card_deck.loc[[card_1,card_2,card_3],:])
        return True

def generate_list(current_board):
    possible_sets = []
    for i in range(len(current_board)):
        for j in range(i+1,len(current_board)):
            for k in range(j+1,len(current_board)):
                possible_sets.append([current_board[i],current_board[j],current_board[k]])
    return possible_sets

def check_list(possible_sets):
    for i in possible_sets:
        value = check_set(i[0],i[1],i[2])
        if value == True:
            current_board.remove(i)
            current_board.remove(j)
            current_board.remove(k)
            break
    return current_board,value

def check_minimum_number(card_deck,deck_cards,current_board):
  new_addition = random.sample(deck_cards,1)
  current_board.append(new_addition[0])

  # print(current_board)
  deck_cards.remove(new_addition[0])
  possible_sets = generate_list(current_board)
  count_sets = 0
  for i in possible_sets:
    value = check_set(i[0],i[1],i[2])
    if value == True:
      # print(len(current_board))
      return (len(current_board))

  b = check_minimum_number(card_deck,deck_cards,current_board)
  return b
```
